```python
g = 9.81
from scipy.integrate import solve_ivp
from math import cos,sin
import matplotlib.pyplot as plt
import numpy as np

def f(t, y): 
    """
    #t float/int 
    y = list 4 float/int
    """
    return [y[1],0,y[3],-g]

def trajectoire(h,v,teta,T): 
    return solve_ivp(f,[0,T],[0,cos(teta)*v,h,sin(teta)*v],max_step=0.01)


teta=np.linspace(-1.57,1.57,50)
for i in teta:
    a=trajectoire(50,10,i,5)
    plt.plot(a.y[0],a.y[2])
#plt.plot(np.linspace(0,140,1000),np.linspace(0,0,1000))
```
