```python
from numpy import exp
import numpy as np 
import matplotlib.pyplot as plt

def f(x): 
    return x*(1-x)
def y(x):
    return 1/(1 + (-1+1/0.5)*exp(-x))
def Euler(T,y0,N): 
    """
    T réel positif 
    y0 réel dans (0,1)
    N Entier strict positif
    """
    X0 = 0 
    h = T/N 
    tab1 = [X0] 
    tab2 = [y0]
    for i in range(1,N): 
        tab1 += [h*i]
        tab2 += [tab2[-1] + (-tab1[-2]+tab1[-1])*f(tab2[-1])]
    return(np.array(tab1),np.array(tab2))

"""
print(Euler(2.5,0.5,10))
print(y(2.5))"""
```

```python
T = 10
y0 = 1/2
N = 100
X = np.linspace(0,T,N)
plt.plot(X,y(X))
plt.plot(X,Euler(T,y0,N)[1])

def Erreur(A,B): 
    return max(abs(A-B)), abs(A-B)
E = Erreur(y(X),Euler(T,y0,N)[1])
plt.plot(X,E[1]*100,label='Erreur zoom x100')
plt.plot(X,np.array([E[0]*100 for i in range(N)]),label=f'{round(E[0],5)}x100')
plt.legend()
```
