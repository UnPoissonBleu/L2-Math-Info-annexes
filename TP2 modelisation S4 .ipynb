```python
from math import exp
import numpy as np 
import matplotlib.pyplot as plt
def f(x): 
    return x*exp(x)-1
def f_(x): 
    return exp(x) + x*exp(x)
def A(x): 
    return -1/f_(x)
def g(x): 
    return x + A(x)*f(x)


def dichotomie(a,b,x,n):
    ai = a 
    bi = b 
    for i in range(n):

        if f((bi+ai)/2)>x : 
            bi = (bi+ai)/2
        elif f((bi+ai)/2)<x:
            ai = (bi+ai)/2
        elif ai == bi == x : 
            return ai,bi
            

    return ai , bi

def dichotomie2(a,b,x,n):
    ai = a 
    bi = b 
    A = [ai]
    B = [bi]
    for i in range(n):
        
        if f((bi+ai)/2)>x : 
            bi = (bi+ai)/2
        elif f((bi+ai)/2)<x:
            ai = (bi+ai)/2
        elif ai == bi == x : 
            return ai,bi
        A.append(ai)
        B.append(bi)

    return A,B

def res(e): 
    i = 0 
    while True:
        i+= 1 
        a,b=dichotomie(0,1,0,i)
        if i > 100_000_000 : 
            return 'error'
        elif b-a < e : 
            return i
```

```python
print(dichotomie(0,1,0,101))
print(res(0.0001))
```
