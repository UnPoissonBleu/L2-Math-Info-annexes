```python
def suite(n):
    
    un = 1/2
    L = [un]
    for i in range(n):
        un = f(un)
        L.append(un)
    return L    
    
print(suite(10))    
```

```python
def visualisation(n): 
    Un = suite(n)
    doubles = [valeur for valeur in Un for i in range(2)]
    plt.figure(figsize = (8, 7))
    plt.plot(doubles[0:-1], doubles[1:], marker="o", markersize = 5)
    plt.plot(x,y2)
    plt.plot(x,y1)
visualisation(100)
```

```python

def fibo(a,b,n):
    """
    return liste
    """
    Un=a
    Un1 = b 
    L = [Un,Un1]
    for i in range(n):
        c = Un1
        Un1 = Un1 + Un 
        Un = c
        L.append(Un1)
        
    return L

a=fibo(1,(1-5**0.5)/2,10)
print(a)
```

```python
def interaction(L0, R0,n): 
    L = [L0]
    R = [R0]
    for i in range(n-1):
        L.append(L[i]*1.2 - R[i]*0.1)
        R.append(0.9*R[i]+0.2*L[i])
    return L,R

a=interaction(2,1,100)
L = np.array(a[0])
R = np.array(a[1])
print(L,R)
xn = L/(L+R)
yn = R/(L+R) 
o = np.linspace(1,10,100)

plt.plot(o,xn)
plt.plot(o,yn)
```
