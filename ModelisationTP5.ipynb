```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.optimize import golden
```

```python
def f(t,y): 
    """
    y : (@,@')
    """
    return y[1], -np.sin(y[0])
```

```python
b = f(1,(0,2) )
print(type(b))
```

```python
def solution(theta_0,theta_prime_0,T):
    solution = solve_ivp(f,[0,T],(theta_0,theta_prime_0),t_eval=np.arange(0,T,1e-4),method='Radau' )
    return solution.y[0], solution.y[1], solution.t
```

```python
theta_0 = 1
theta_prime_0 = 0
T = 15
def test(theta_0,theta_prime_0,T):
    return solution(theta_0,theta_prime_0,T)
sol = solution(theta_0,theta_prime_0,T)
plt.plot(sol[2],sol[0])
plt.plot(sol[2],sol[1])
print(sol[2][list(sol[0]).index( max(abs(sol[0])) ) ],"est le temps o√π la vitesse est maximale ")

print(max(sol[0]))
```
