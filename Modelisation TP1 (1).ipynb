```python
#4
import numpy as np 
import matplotlib.pyplot as plt

def suite_recurrente(f,a0,n):
    a = [a0]
    for i in range(n): 
        a += [f(a[-1])]
    return np.array(a)
```

```python
def g(x):
    return np.sqrt(x+2)
def h(x): 
    return x**2 

n = 10 
a0s = np.linspace(0,4,9)
for i in a0s : 
    plt.plot(suite_recurrente(g,i,n),label=f'a0={i}')
plt.legend(bbox_to_anchor=(1, 1.0),prop={'size': 10}) 
plt.title('(Un)')
plt.plot()
```

```python
n = 10 
a0s = np.linspace(0,.991,5)
for i in a0s : 
    plt.plot(suite_recurrente(h,i,n),label=f"a0 = {i}")
plt.legend(bbox_to_anchor=(1,1),prop={"size":10})
plt.title("(Vn)")
plt.plot()
```

```python
n = 200
a0s = np.linspace(0,.75,5)
for i in a0s : 
    plt.plot(suite_recurrente(np.sin,i,n),label=f"a0 = {i}")
plt.legend(bbox_to_anchor=(1,1),prop={"size":10})
plt.title("(Wn)")
plt.plot()
```

```python
n=100000
a0s=[0,0.25,0.5,0.75]
for a0 in a0s: 
    plt.plot(suite_recurrente(np.sin,a0,n),label=f'a0={a0}')
plt.legend()
plt.title('La suite w')
plt.plot();
```
