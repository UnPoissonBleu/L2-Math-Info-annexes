```python
#4
from math import * 
import numpy as np 
import matplotlib.pyplot as plt

def f(x):
    return x**2
def _(f,a,n):
    L = [a] 
    for i in range(n):
        L.append(f(L[i]))
    return np.array(L)
print(_(f,1.1,5))
```

```python
def g(x):
    return sqrt(x+2)
def h(x): 
    return x**2 

def colimacon(f,u0,n,x,y): 
    L = [u0]
    for i in range(n): 
        L += [f(L[-1])]

    U = L
    doubles = [valeur for valeur in U for i in range(2)]
    plt.figure(figsize = (8, 7))
    plt.plot(doubles[0:-1], doubles[1:], marker="o", markersize = 5)
    plt.plot(x,y,label='y=f(x)')
    plt.plot(x,x,label='y=x')
    plt.legend()
    plt.plot(u0,u0,color='red',marker ='o')
    return(U[-1])

X = np.linspace(0,2,100)
Yg = np.array([g(e) for e in X])
Yh = np.array([h(e) for e in X])

#plt.plot(X,Yg)
#plt.plot(X,X,color='red') 
#plt.plot(X,Yh)
colimacon(g,1,100,X,Yg)
plt.plot(np.array([2 for i in range(10)]), np.linspace(0,4,10),color='black' )
colimacon(h,0.5,100,X,Yh)
plt.plot(np.array([1 for i in range(10)]), np.linspace(0,4,10),color='black' )


```

```python
X = np.linspace(-1.5,1.5,100)
Ysin = np.array([sin(e) for e in X])
#plt.plot(X,Ysin)
#plt.plot(X,X,color='red')
colimacon(sin,1,100,X,Ysin)
```
