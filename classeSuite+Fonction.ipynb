```python
class Sequence: 
    def __init__(self,f,n0):
        """ 
        self - the sequences herself
        f : <Class Function> a function that represent the sequence ex : Un = n²+2/n : f(x) = x²+2/x
        n0 the first step of ur sequence, f(n) must exist for n>n0
        """
        self.function=f 
        self.first_term = self.function(n0)
        self.first_step = n0
        
        
    def __str__(self): 
        return f"{self.function} | n0 = {self.first_step}, f({self.first_step})={self.first_term}"
    
    def __call__(self,n): 
        return self.function(n)
    
    def __add__(self,other):
        return Sequence(self.function + other.function,max(self.first_step,other.first_step))
    
        
    
class Function: 
    def __init__(self,name,f,expression): 
        """ 
        f is a lambda function 
        expression is is litteral expression
        """
        self.name = name
        self.function = f 
        self.expression = expression

    def __str__(self):
        return f"{self.expression}" 
    
    def __call__(self,x): 
        return self.function(x)
    
    def __add__(self,other): 
        return Function(f"{self.name}+{other.name}",lambda x: self.function(x)+other.function(x),f"{self.expression}+{other.expression}")

    def __mul__(self,other): 
        return Function(f"{self.name}*{other.name}",lambda x: self.function(x)*other.function(x),f"({self.expression})({other.expression})")
f = Function("f",lambda x:x**2,"x**2")
g = Function("g",lambda x:x,"x")
print((f+g)(5))

U = Sequence(Function('f',lambda x:x**2,"n**2"),3)
V = Sequence(Function('f',lambda x:2*x,"2n"),4)
print(U+V)
print(U)
print(U(10))
```
